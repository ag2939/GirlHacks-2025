```python
import re

# Combined regex patterns for different units
pattern = re.compile(
    r"(^package\s+[\w.]+;)|"  # package
    r"(^import(\s+static)?\s+[\w.]+(\.[\w*]+)?;)|"  # imports
    r"(^public\s+class\s+\w+(\s+extends\s+\w+)?(\s+implements\s+[\w, ]+)?\s*\{?)|"  # class
    r"(^\s*(private|public|protected)\s+[\w<>[\],\s]+;\s*$)|"  # fields
    r"((^\s*@\w+(\([^)]*\))?\s*\n\s*)*(public|private|protected)?\s*[\w<>\[\]]+\s+\w+\s*\([^)]*\)\s*\{?)|"  # methods with optional annotations
    r"(^\s*@\w+(\([^)]*\))?\s*$)|"  # standalone annotations
    r"(^\s*}\s*$)",  # closing brace
    re.MULTILINE
)

def split_java_units(code_lines):
    code_text = "\n".join(code_lines)
    matches = list(pattern.finditer(code_text))
    units = []

    for match in matches:
        units.append(match.group(0).strip())

    return units


units = split_java_units(lines)

print("=== Split Units ===")
for i, unit in enumerate(units, 1):
    print(f"{i}. {unit}")
```

```python
import re

def split_java_sections(code_lines):
    sections = []
    buffer = []
    brace_count = 0
    in_block = False

    # Combine lines to preserve formatting
    lines = code_lines[:]

    for line in lines:
        stripped = line.strip()

        # Always include package/import lines as separate units
        if re.match(r"^package\s+[\w.]+;|^import(\s+static)?\s+[\w.]+(\.[\w*]+)?;", stripped):
            sections.append(stripped)
            continue

        # Detect start of a class or method block
        if re.match(r"^public\s+class\s+\w+.*\{", stripped) or \
           re.match(r"^(public|private|protected)?\s*[\w<>\[\]]+\s+\w+\s*\([^)]*\)\s*\{", stripped) or \
           re.match(r"^\s*@\w+", stripped):  # annotation
            buffer.append(line)
            brace_count += line.count("{") - line.count("}")
            in_block = True
            continue

        # If inside a block, accumulate lines
        if in_block:
            buffer.append(line)
            brace_count += line.count("{") - line.count("}")
            if brace_count == 0:
                sections.append("\n".join(buffer).strip())
                buffer = []
                in_block = False
            continue

        # Otherwise, treat as standalone (fields, etc.)
        if stripped:
            sections.append(stripped)

    return sections


if __name__ == "__main__":
    java_code_lines = [
        "package com.example.shiksharemastered;",
        "import static java.lang.String.valueOf;",
        "import androidx.appcompat.app.AppCompatActivity;",
        "public class UploadExperiments extends AppCompatActivity {",
        "    private DatabaseReference myRef;",
        "    @Override",
        "    protected void onCreate(Bundle savedInstanceState) {",
        "        super.onCreate(savedInstanceState);",
        "    }",
        "}"
    ]

    units = split_java_sections(java_code_lines)
    for i, unit in enumerate(units, 1):
        print(f"--- Unit {i} ---")
        print(unit)
        print()
```
